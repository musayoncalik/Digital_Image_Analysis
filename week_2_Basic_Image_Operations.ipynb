```python
# Temel Görüntü İşlemleri (BASIC IMAGE OPORATIONS)
# VEYA Görüntü Okuma, Görüntü Gösterme ve Piksel Düzeyinde İşlemler
```

```python
# Bu konu başlığı altında yapılan işlemler:
	# 1. Görüntü okuma (image reading) – cv2.imread()
	# 2. Görüntü özelliklerini öğrenme – boyut (height, width, channels)
	# 3. Renk uzayı dönüşümü – BGR → RGB → Grayscale
	# 4. Görüntü kaydetme (image writing) – cv2.imwrite()
	# 5. Piksel değerlerini inceleme (max, min)
	# 6. Piksel normalizasyonu (0–255 → 0–1 aralığı)
	# 7. Görüntülerin matplotlib ile görselleştirilmesi (subplot, cmap)

# Bir resmi okur → boyutlarını inceler → RGB ve griye çevirir → görüntüler → kaydeder 
# → piksel değerlerini normalize eder.
```

```python
import cv2               # cv2: OpenCV → görüntü okuma, yazma, dönüştürme
import numpy as np       # numpy: sayısal işlemler (piksel değerleriyle oynamak)
import matplotlib.pyplot as plt  # matplotlib: görselleştirme (resimleri göstermek için)
import os
```

```python
os.getcwd()
```

```python
#cv2.imread(dosya_yolu_adı.uzantısı,0) => görüntüyü gri seviyede (tek kanallı oku)
#cv2.imread(dosya_yolu_adı.uzantısı,1) => görüntüyü rgb (renkli) seviyede (3 kanallı oku)
#cv2.imread(dosya_yolu_adı.uzantısı) => görüntüyü rgb (renkli) seviyede (3 kanallı oku)

my_img = cv2.imread("../image/Lion_waiting_in_Namibia.jpeg")
```

```python
type(my_img)
```

```python
print(my_img.shape)  # Görüntü boyutlarının öğrenilmesi
```

```python
# # Görüntü boyutlarını öğrenmek .shape()

print(f"Görüntünün yüksekliği = {my_img.shape[0]}")
print(f"Görüntünün genişliği = {my_img.shape[1]}")
print(f"Görüntünün kanal sayısı = {my_img.shape[2]}")
```

```python
# RGB (Red-Green-Blue), bilgisayarda renkle 0 ile 255 arasında temsil edilir
# Yani, renkler gösterilirken uint8 veri tipi kullanılır.
# uint8 demek unsignedint8 demektir. İşaretsiz 8 bitlik değer:
# 0 ile 255 arasında (her ikisi de dahil) 256 adet tamsayı vardır.
```

```python
# RGB ile Gray level (gri seviye) görüntü arasında sadece kanal ve gösterim
# farkı vardır. Kanal farkı, gri seviye görüntüde 1 kanal vardır, RGB'de ise 3 tane.
# RGB, renkli görüntü gösterirken, gri seviye sadece siyah veya beyaz tonlarında
# renkleri gösterir.
```

```python
# cv2 ile BGR kanal sıralamasında okunan görüntüyü RGB kanal sıralamasına dönüştürüyoruz
# BGR → RGB ve Griye dönüştürme
img_rgb_1 = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
```

```python
# Görüntülerin gösterilmesi
plt.imshow(img_rgb_1)
plt.axis("off")
```

```python
# Renkli (RGB) görüntüyü gri tonlamalı (grayscale) görüntüye dönüştürür
img_gray_1 = cv2.cvtColor(img_rgb_1, cv2.COLOR_RGB2GRAY)
```

```python
# Gri tonlamalı görüntüyü matplotlib ile gösterir ve eksenleri gizler
plt.imshow(img_gray_1, cmap="gray")
plt.axis("off")
```

```python
# Renkli görüntünün boyutlarını (yükseklik, genişlik, kanal sayısı) ekrana yazdırır
print(img_rgb_1.shape)

# Gri tonlamalı görüntünün boyutlarını (yükseklik, genişlik) ekrana yazdırır
print(img_gray_1.shape)
```

```python
# RGB görüntüde (500,500) koordinatındaki pikselin R, G, B değerlerini ekrana yazdırır
print(f"RGB görüntünün piksel değeri = {img_rgb_1[500,500]}")

# Gri tonlamalı görüntüde (500,500) koordinatındaki tek kanal piksel değerini ekrana yazdırır
print(f"Gri görüntünün piksel değeri = {img_gray_1[500,500]}")
```

```python
# Görüntünün Yüksekliği satır sayısıyla doğru orantılıdır
# Görüntünün genişliği sütun sayısıyla doğru orantılıdır
```

```python
# plt.subplot(nrows, ncols, index) şeklinde kullanılır.
# nrows → toplam satır sayısı
# ncols → toplam sütun sayısı
# index → hangi hücreye çizim yapılacağını belirtir (soldan sağa, üstten alta)

# plt.subplot(1,2,1)
# 1 satır, 2 sütunluk tablo içinde **ilk hücreyi aktif et” ve buraya grafiği çiz
```

```python
# Tek bir plot içerisinde alt plotlarla (subplot) birden fazla görüntünün
# gösterimi

fig = plt.figure(figsize=(10,7))

# Subplot ile yan yana gösterim
#plt.subplot(satır_sayısı,sütun_sayısı,o anki gösterilecek görüntü)
plt.subplot(1,2,1)
plt.imshow(img_rgb_1)
plt.axis("off")
plt.title("RGB Image")

plt.subplot(1,2,2)
plt.imshow(img_gray_1,cmap="gray")
plt.axis("off")
plt.title("Gray Level Image")


plt.tight_layout()
plt.savefig("My_image_with_subplot.png",dpi=300)
plt.show()
```

```python
# Görüntünün kaydedilmesi: Görüntü dosya olarak bilgisayara kaydedilir.
cv2.imwrite("my_rgb_image.png",my_img)
```

```python
# Piksel değerlerinin incelenmesi
# Görüntüdeki en parlak (max) ve en karanlık (min) piksel değerlerini bulur.
#Genelde 0 (siyah) – 255 (beyaz) aralığındadır.

print(f"Görüntüde yer alan en yüksek piksel değeri = {np.max(my_img)}")
print(f"Görüntüde yer alan en düşük piksel değeri = {np.min(my_img)}")
```

```python
# Görüntüde Piksel Normalizasyon
# En düşük piksel değeri 0 ve en yüksek piksel değeri 255 olduğu için
# Piksel değerlerini 0–255 aralığından 0–1 aralığına dönüştürür.

img_rgb_2 = img_rgb_1 / 255.0
```

```python
plt.imshow(img_rgb_2)
plt.axis("off")
```

```python
# RGB görüntüdeki en yüksek piksel değerini ekrana yazdırır
print(f"Görüntüde yer alan en yüksek piksel değeri = {np.max(img_rgb_1)}")

# RGB görüntüdeki en düşük piksel değerini ekrana yazdırır
print(f"Görüntüde yer alan en düşük piksel değeri = {np.min(img_rgb_1)}")

# Görsel ayırıcı çizgi
print("-"*50)

# Normalize edilmiş görüntüdeki en yüksek piksel değerini ekrana yazdırır
print(f"Normalize görüntüde yer alan en yüksek piksel değeri = {np.max(img_rgb_2)}")

# Normalize edilmiş görüntüdeki en düşük piksel değerini ekrana yazdırır
print(f"Normalize görüntüde yer alan en düşük piksel değeri = {np.min(img_rgb_2)}")
```
